# 텍스트 특징 추출 (TF-IDF)

### 1. 샘플 코퍼스로 연습

```python
sample_corpus = [
    '자연어처리 강의를 시작하겠습니다.',
    '자연어처리는 재미있습니다.',
    '밥을 먹고 강의를 듣고 있습니다.',
    '이번 자연어처리 강의는 한국어 자연어처리입니다.'
]
```

```python
import sklearn
```

```python
from konlpy.tag import Okt
def my_tokenizer(text):
    return Okt().nouns(text)
```

```python
# TfidfVectorizer 객체 생성
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(tokenizer=my_tokenizer)
# from sklearn.feature_extraction.text import CountVectorizer
# vectorizer = CountVectorizer(tokenizer=my_tokenizer)
# 특징 집합과 관련 데이터 모델을 생성
vectorizer.fit(sample_corpus)
print(vectorizer.get_feature_names_out())
# 특징 벡터 추출
sample_dtm = vectorizer.transform(sample_corpus)
print(type(sample_dtm))
sample_dtm.toarray()
```

### 2. 다음 영화 리뷰로 연습

```python
from konlpy.tag import Okt
def my_tokenizer(text):
    t = Okt()
    # 명사(일반 명사, 고유 명사), 동사, 형용사 단어 추출
    my_tags = ["Noun", "Verb", "Adjective"]

    # 불용어 삭제 : 의미없는데 고빈도로 나오는 단어 추가
    my_stopwords = ["내", "내내", "티", "나", "들인건", "할수밖에", "없다", "보는", "정말", "하는", "보고", "입니다", "그냥", "정도", "해서", "있는", "봤는데", "거", "것", "그", "꼭", "더", "듯", 
                    "때", "뭐", "볼", "수", "안", "왜", "이", "점", "좀", "편"]
    return [word for word, tag in t.pos(text) if tag in my_tags and word not in my_stopwords and len(word) > 1]
```

```python
### 1. 데이터 준비
import pandas as pd

datafile = './data/daum_movie_review.csv'
data_df = pd.read_csv(datafile, usecols=['review'])
reviews = list(data_df.review)
print(reviews[:3])
```

```python
# TfidfVectorizer 객체 생성
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(
    tokenizer = my_tokenizer,
    max_features = 50
    )
# 특징 집합과 관련 데이터 모델을 생성
vectorizer.fit(reviews)
print(vectorizer.get_feature_names_out())
# 특징 벡터 추출
sample_dtm = vectorizer.transform(reviews)
print(type(sample_dtm))
sample_dtm.toarray()
```
