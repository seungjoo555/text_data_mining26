```python
class Account:
    # balance = 100

    def __init__(self):     # 멤버 변수 balance 초기화 -> 변수 선언
        self.balance = 0

    def deposit(self, amount):
        self.balance += amount
    
    def withdraw(self, amount):
        self.balance -= amount
```

```python
a1 = Account()
a1.deposit(100)
print(a1.balance)
a1.withdraw(50)
print(a1.balance)

print("a1 :", a1.balance)
# print("Account:", Account.balance)

# print(Account.balance.__hash__)
print(a1.balance.__hash__)
```

```python
class Account:
    def __init__(self):
        self.__balance = 0

    def deposit(self, amount):
        self.__balance += amount
    
    def withdraw(self, amount):
        self.__balance -= amount
    
    def get_balance(self):
        return self.__balance
```

```python
a1 = Account()
a1.deposit(100)
a1.withdraw(50)
# print("a1 :", a1.__balance)
print("a1 :", a1.get_balance())
```
